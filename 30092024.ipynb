```python
# Definir os parametros de conexao do mysql
host = '35.225.237.49' # IP PUBLICO DA SUA INSTANCIA DO MYSQL
user = 'root' # USUARIO
password = '' # Senha
database = 'locadora'
```

```python
# Função que ao ser chamada irá realizar a nossa consulta SQL, toda vez que
# For chamada passaremos a 'query' no parametro
def sqlQuery(consulta):
  connection = mc.connect(
    host=host,
    user=user,
    password=password,
    database=database
  ) # criar a conexão

  cursor = connection.cursor() # criar o cursor que vai realizar as acoes

  cursor.execute(consulta) # realizar a consulta

  result = cursor.fetchall() # guardar o resultado

  df = pd.DataFrame(result, columns=cursor.column_names) # criar o DataFrame

  cursor.close() # Encerrar o cursor

  connection.close() # Encerrar a conexão

  return df # retorna para fora da função o Dataframe resultante
```
